#Doing hupothesis for stats comparision

H0 = Mean of Train and Test data are same

Ha= Mean of train and test data are not sam

```python
#randomly selecting feature 2
from statsmodels.stats import weightstats as stests

ztest,pval = stests.ztest(X_test['CASING'], x2=X_train["CASING"],value=0,alternative="two-sided")
print("Pvalue of original and Train data",pval)
if(pval<0.05):
    print("Reject Null Hypothesis, Mean of Test and train data for feature CASING is not same") 
else:
    print("Fail to reject Null Hypothesis,Mean of test and train data for feature CASING is same")
```

```python
#Anova randomly selecting feature 4

#let my H0 = MEan of original, train and test data are same 
#Ha Mean of original, train and test data are not same 

from scipy.stats import f_oneway
ztest,pval= f_oneway(df_CPM["IMPELLER"],X_train["IMPELLER"],X_test["IMPELLER"])
print("feature 0, Pvalue", pval)
if(pval<0.05):
    print("Reject Null Hypothesis,Mean of original, train and test data are same")
else:
    print("Fail to reject Null Hypothesis,Mean of original, train and test data are same")
```
